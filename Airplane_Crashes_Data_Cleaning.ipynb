```python
# Qn: Create a new dataframe from the raw/uncleaned data called 'fatality_location' with the following columns: 'Date', 'Location', 'Aboard', 'Fatalities'.
# Solution approach: We will create a new DataFrame called 'fatality_location' by selecting only the specified columns ('Date', 'Location', 'Aboard', 'Fatalities') from the original uncleaned DataFrame. This allows us to focus on the key variables related to fatalities and their locations while retaining the original data for reference. We will use double brackets to ensure that the resulting object is a DataFrame, which is important for maintaining the structure and functionality of the data for further analysis.

print("=" * 60)
print("QUESTION 5: Create Filtered DataFrame - 'fatality_location'")
print("=" * 60)

print("\nCreating 'fatality_location' dataframe...")
print(f"Source data shape: {df.shape}")

# Select only the specified columns from the original dataframe
# Using double brackets to maintain dataframe structure (not Series)
fatality_location = df[['Date', 'Location', 'Aboard', 'Fatalities']]

print(f"New dataframe shape: {fatality_location.shape}")
print(f"Columns in new dataframe: {list(fatality_location.columns)}")

# Displaying  information about the new dataframe
print("\n" + "-" * 60)
print("FATALITY_LOCATIONS DATAFRAME - FIRST 10 ROWS")
print("-" * 60)
print(fatality_location.head(10))

print("\n" + "-" * 60)
print("FATALITY_LOCATIONS DATAFRAME - INFO")
print("-" * 60)
print(fatality_location.info())

print("\n" + "-" * 60)
print("FATALITY_LOCATIONS DATAFRAME - MISSING DATA")
print("-" * 60)
missing_in_fatality = pd.DataFrame({
    'Column': fatality_location.columns,
    'Missing_Count': fatality_location.isnull().sum(),
    'Missing_Percentage': (fatality_location.isnull().sum() / len(fatality_location) * 100).round(2)
})
print(missing_in_fatality)

print("\n" + "-" * 60)
print("FATALITY_LOCATIONS DATAFRAME - BASIC STATISTICS")
print("-" * 60)
print(fatality_location.describe())

print(f"\nNew 'fatality_location' dataframe created successfully with columns: {fatality_location.columns.tolist()}")

```

```python
# QN : Determine the date of the highest number of recorded fatalities from this new dataframe.
# Solution approach: We will use the idxmax() method to find the index of the row with the maximum number of fatalities in the 'Fatalities' column of the 'fatality_location' DataFrame. This method is efficient for identifying the position of the maximum value without needing to sort the entire DataFrame. Once we have the index, we can use it to extract the corresponding row, which will contain the date and location of the crash with the highest recorded fatalities. This approach allows us to quickly and accurately identify the most severe crash in terms of fatalities.
print("=" * 60)
print("QUESTION 6: Date of Highest Recorded Fatalities")
print("=" * 60)

# Find the index (position) of the row with maximum fatalities
# idxmax() returns the label/index of the maximum value, which is more efficient than
# sorting the entire dataframe when we only need one maximum value
max_fatalities_index = fatality_location['Fatalities'].idxmax()

# Extract the entire row with maximum fatalities
max_fatalities_row = fatality_location.loc[max_fatalities_index]

print(f"\nFATALITY RECORD WITH HIGHEST DEATH TOLL:")
print("-" * 60)
print(max_fatalities_row)



# Extract individual values for clarity
date_highest_fatalities = max_fatalities_row['Date']
location_highest = max_fatalities_row['Location']
aboard_highest = max_fatalities_row['Aboard']
fatalities_highest = max_fatalities_row['Fatalities']

print("\n" + "-" * 60)
print("ALTERNATIVE VERIFICATION METHOD:")
print("-" * 60)

# Verify using an alternative approach - sorting by Fatalities in descending order
# This helps us see the top crashes by fatality count
print("\nTop 5 crashes by fatality count:")
top_5_crashes = fatality_location.nlargest(5, 'Fatalities')[['Date', 'Location', 'Fatalities']]
print(top_5_crashes)

print("\n" + "=" * 60)
print(f"CONCLUSION: The highest number of recorded fatalities ({int(fatalities_highest)}) occurred on {date_highest_fatalities}")
print(f"Location: {location_highest}")
print("=" * 60)

```

```python
# Qn: Compare the number of passengers aboard to the number of fatalities per crash. Are there any recorded crashes where there were no fatalities? If so, how many?
# Solution approach: We will create a new column to calculate the fatality rate (fatalities/aboard) to compare the severity of crashes. We will also check for any rows where fatalities are zero to identify no-fatality crashes. This analysis helps us understand the relationship between the number of people aboard and the number of fatalities, and whether there are instances of crashes with no loss of life.
print("=" * 60)
print("QUESTION 7: Aboard vs Fatalities Comparison")
print("=" * 60)

# Use only rows where both values exist so calculations are reliable.
comparison_df = fatality_location[['Date', 'Location', 'Aboard', 'Fatalities']].dropna(subset=['Aboard', 'Fatalities']).copy()

# Fatality rate helps compare crashes of different aircraft sizes.
comparison_df['Fatality_Rate_%'] = (comparison_df['Fatalities'] / comparison_df['Aboard']) * 100
comparison_df['Survivors'] = comparison_df['Aboard'] - comparison_df['Fatalities']

# Overall Comparison Summary...
print("\nOverall comparison summary:")
print("-" * 60)
print(f"Total crashes analyzed (non-missing Aboard & Fatalities): {len(comparison_df)}")
print(f"Average people aboard per crash: {comparison_df['Aboard'].mean():.2f}")
print(f"Average fatalities per crash: {comparison_df['Fatalities'].mean():.2f}")
print(f"Average fatality rate: {comparison_df['Fatality_Rate_%'].mean():.2f}%")

# Crashes with no fatalities are rows where Fatalities equals 0.
no_fatality_crashes = comparison_df[comparison_df['Fatalities'] == 0]
no_fatality_count = len(no_fatality_crashes)

print("\n" + "-" * 60)
if no_fatality_count > 0:
    print(f"Yes. There are {no_fatality_count} recorded crashes with zero fatalities.")
    print("Full List of no-fatality crashes:")
    print(no_fatality_crashes[['Location', 'Aboard', 'Fatalities']])
else:
    print("No. There are no recorded crashes with zero fatalities.")

```

```python
# QN: In the ‘fatality_locations’ dataframe, the location column has data items represented as ‘Region, U.S. State/Country’. Split this column to have two columns of regions and states/countries.
# Solution approach: We will use the str.rsplit() method to split the 'Location' column into two new columns: 'Region' and 'State_Country'. The rsplit() method allows us to split from the right, which is important because some regions may contain commas. By specifying n=1, we ensure that only the last comma is used for splitting, preserving the integrity of the region names. We will also use expand=True to directly create two separate columns from the split operation. This approach ensures that we accurately separate the region and state/country information while handling any potential complexities in the location data.
print("=" * 60)
print("QUESTION 8: Split Location into Region and State/Country")
print("=" * 60)

# The earlier cells created `fatality_location` (singular). We keep that as the source.
# A copy is used so the original filtered dataframe remains unchanged.
fatality_locations = fatality_location.copy()

# Reasoning:
# 1) Location values are in the form "Region, State/Country".
# 2) We split on the LAST comma, not the first, so regions that already contain commas
#    still remain intact in the Region part.
# 3) expand=True returns two separate columns directly.
location_split = fatality_locations['Location'].str.rsplit(',', n=1, expand=True)

# Assign cleaned split columns.
fatality_locations['Region'] = location_split[0].str.strip()
fatality_locations['State_Country'] = location_split[1].str.strip()

print("\nPreview after splitting Location:")
print(fatality_locations[['Location', 'Region', 'State_Country']].head(10))

print("\nMissing values after split:")
print(fatality_locations[['Region', 'State_Country']].isnull().sum())

print("\nDone: 'fatality_locations' now has separate 'Region' and 'State_Country' columns.")

```

```python
#QN: Order the dataframe by the number of fatalities from highest to lowest and select the first 100.
# Solution approach: We will use the sort_values() method to sort the 'fatality_locations' DataFrame by the 'Fatalities' column in descending order (highest to lowest). After sorting, we will use the head(100) method to select the top 100 rows from the sorted DataFrame. This approach allows us to efficiently identify and analyze the most severe crashes in terms of fatalities, providing insights into the locations and circumstances of these incidents.
print("=" * 60)
print("QUESTION 9: Top 100 Crashes by Fatalities")
print("=" * 60)

# Used the dataframe created in the previous cell (contains Region and State_Country split).
# Sort by Fatalities in descending order so the most severe crashes appear first.
top_100_fatalities = fatality_locations.sort_values(by='Fatalities', ascending=False).head(100)

print(f"\nTop 100 dataframe shape: {top_100_fatalities.shape}")
print("=" * 60)
# The Top 100
print("The Top 100")
print(top_100_fatalities[['Date', 'Location', 'Fatalities']])

# Display the full selected 100 rows in notebook output.
top_100_fatalities

```

```python
# QN: Generate a pie chart that shows the distribution of the top 25 number of fatalities per country/U.S. State. [Hint: charts can be generated from the seaborn library]
# Solution approach: We will first prepare the data by aggregating the total number of fatalities for each country/U.S. state from the 'State_Country' column in the 'fatality_locations' DataFrame. We will then select the top 25 countries/states with the highest total fatalities. Finally, we will use matplotlib to create a pie chart that visualizes the distribution of fatalities among these top 25 locations. The pie chart will include percentage labels and a legend for clarity, allowing us to easily compare the relative impact of crashes across different regions.
print("=" * 60)
print("QUESTION 10: Pie Chart - Top 25 Fatalities by Country")
print("=" * 60)

# Prepare clean chart data: keep rows where both State_Country and Fatalities exist.
chart_data = fatality_locations[['State_Country', 'Fatalities']].dropna().copy()
chart_data['State_Country'] = chart_data['State_Country'].str.strip()
chart_data = chart_data[chart_data['State_Country'] != ""]

# Aggregate total fatalities by country/state, then take the top 25 highest totals.
fatalities_by_state_country = (
    chart_data.groupby('State_Country', as_index=False)['Fatalities']
    .sum()
    .sort_values(by='Fatalities', ascending=False)
)
top_25_fatalities = fatalities_by_state_country.head(25)

print(f"Rows used for charting: {len(chart_data)}")
print(f"Unique countries/states found: {fatalities_by_state_country['State_Country'].nunique()}")
print("\nTop 25 countries/states by total fatalities:")
print(top_25_fatalities)

# Used seaborn color palette with matplotlib pie chart for clear, consistent styling.
colors = sns.color_palette('tab20', n_colors=len(top_25_fatalities))

plt.figure(figsize=(12, 12))
wedges, texts, autotexts = plt.pie(
    top_25_fatalities['Fatalities'],
    labels=None,  # Put labels in legend to reduce overlap.
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    pctdistance=0.8,
    textprops={'fontsize': 9}
)
# Add a circle at the center to make it a donut chart for better readability.
plt.title('Distribution of Top 25 Number of Fatalities per Country', fontsize=14)
plt.legend(
    wedges,
    top_25_fatalities['State_Country'],
    title='Country / States',
    loc='center left',
    bbox_to_anchor=(1.02, 0.5),
    fontsize=8
)
plt.tight_layout()
plt.show()

```
